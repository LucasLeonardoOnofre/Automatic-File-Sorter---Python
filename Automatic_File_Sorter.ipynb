```python
# Import required modules
import os, shutil
```

```python
# Define the path to the folder containing files to be sorted
path = r"C:/Users/name/Documents/Projects/Python/Project- Automatic File Sorter in File Explorer/"

```

```python
# Get a list of all files in the folder
file_name = os.listdir(path)
```

```python
# Define the names of the folders where files will be moved
folder_names =['pdf files','txt files','xlsx files']
# Create folders if they do not already exist
for loop in range(0,3):
  if not os.path.exists(path + folder_names [loop]):
     os.makedirs(path + folder_names [loop])
```

```python
# Loop through each file in the folder
for file in file_name:
    if file.endswith(".pdf") and not os.path.exists(os.path.join(path, "pdf files", file)):
        src = os.path.join(path, file)   # Source path of the file
        dst = os.path.join(path, "pdf files", file) # Destination path in 'xlsx files' folder
        shutil.move(src, dst)   # Move the file

    elif file.endswith(".txt") and not os.path.exists(os.path.join(path, "txt files", file)):
        src = os.path.join(path, file)
        dst = os.path.join(path, "txt files", file)
        shutil.move(src, dst)

    elif file.endswith(".xlsx") and not os.path.exists(os.path.join(path, "xlsx files", file)):
        src = os.path.join(path, file)
        dst = os.path.join(path, "xlsx files", file)
        shutil.move(src, dst)
```
